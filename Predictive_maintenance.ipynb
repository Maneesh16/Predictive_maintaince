```python
# dropping these columns because there is no change in these variables 
dfn=df.drop(columns=['setting3','s1','s5','s10','s16','s18','s19','s6'])
```

```python
dfn=dfn.drop(columns=['setting2','setting1'])
```

```python
from sklearn.preprocessing import StandardScaler # importing standard scaler
```

```python
sc=StandardScaler() # using standard scaler to convert the values of the variables with mean = 0,std=0
```

```python
xsc=sc.fit_transform(dfn.drop(columns='ttf'))
```

```python
x=pd.DataFrame(xsc,columns=dfn.drop(columns='ttf').columns)
```

```python
y=dfn['ttf']
```

```python
import statsmodels.api as sm # importing statsmodel
```

```python
for i in range(len(x.keys())):
    model=sm.OLS(y,sm.add_constant(x)).fit()
    print(model.summary(),'\n\n\n')
    print(model.pvalues.sort_values(ascending=False))
    print('$'*100)
    print('\n\n\n')
    if model.pvalues.sort_values(ascending=False)[0]>0.025:
        x.drop(columns=model.pvalues.sort_values(ascending=False).index[0],axis=1,inplace=True)
    else:
        break
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2) # splitting the data into training and testing
```

```python
model2=sm.OLS(y_train,sm.add_constant(x_train)).fit() # fitting the linear regression model using statsmodels
model2.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor # checking correleation between the dependent variables
vif=pd.DataFrame()
vif["Features"]=x.keys()
vif["VIF score"]=[variance_inflation_factor(x.values,i) for i in range(len(x.keys()))]
vif
```

```python
y_pred=model2.predict(sm.add_constant(x_test)) # predicting the y yalues
```
